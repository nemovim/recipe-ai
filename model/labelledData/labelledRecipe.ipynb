```python
import requests
import json
import csv
```

```python
def getRecipe(id):
    res = requests.get(f'https://recipe-ai-olive.vercel.app/api/get?id={id}')
    data = res.json()
    ingArr = data['label']['ingredientArr']
    mainArr = data['label']['mainArr']
    return ingArr
```

```python
for i in range(6900000, 6900001):
    print(getRecipe(i))
```

```python
recipeArr = []
with open('labeling.labels.json', encoding='utf-8') as recipeFile:
    recipeJSON = json.load(recipeFile)
    for i, recipe in enumerate(recipeJSON):

        allArr = recipe['ingredientArr']
        mainArr = list(filter(lambda ing: True if ing != None else False, recipe['mainArr']))

        if mainArr == [-1] or mainArr == []:
            continue

        # onehot_mainArr = map(lambda ing: 1 if ing in mainArr else 0, allArr)
        # recipeArr.append([i, recipe['recipeId'], '/'.join(map(str, allArr)), '/'.join(map(str,onehot_mainArr))])
        recipeArr.append([i, recipe['recipeId'], allArr, mainArr])
```

```python
def updateIng(mappingDict, ingArr):
    ingSet = set()
    for ing in ingArr:
        try:
            ingSet.add(mappingDict[ing])
        except:
            continue
    return list(ingSet)
```

```python
mappingDict = {} 
with open('../../ingredient/idMappings2.csv', encoding='utf-8') as f:
    r = csv.reader(f)
    for mapping in r:
        if mapping[0] == 'old':
            continue
        mappingDict[int(mapping[0])] = mapping[1]
    
print(mappingDict)
print(recipeArr)

for recipe in recipeArr:
    allIng = '/'.join(updateIng(mappingDict, recipe[2]))
    mainIng = '/'.join(updateIng(mappingDict, recipe[3]))
    if mainIng == '':
        print(mainIng)
    recipe[2] = allIng
    recipe[3] = mainIng

print(recipeArr)
```

```python
with open('mainIngredients.csv', '+w', encoding='utf-8', newline='') as mainFile:
    writer = csv.writer(mainFile)
    writer.writerow(['id', 'recipe', 'all', 'main'])
    writer.writerows(recipeArr)
```
