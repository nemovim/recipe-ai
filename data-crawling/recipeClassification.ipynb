```python
import csv
```

```python
def oneHotEncoding(vector):
    ingCnt = 700
    encodedVector = [0 for _ in range(ingCnt)]

    for ing in vector:
        encodedVector[int(ing)] = 1

    return encodedVector 
```

```python
def compress(encodedVector):
    zeroVectorArr = (''.join(map(str, encodedVector))).split('1')
    return ([len(zeros) for zeros in zeroVectorArr])

def decompress(compressedVector):
    return list(map(int, list('1'.join(['0'*i for i in compressedVector]))))
```

```python
def mappingId(fileName, _mappingDict):
    with open(fileName, encoding='utf-8') as f:
        r = csv.reader(f)
        for line in r:
            _mappingDict[line[0]] = line[1]
    return _mappingDict
```

```python
mappingDict1 = mappingId('../ingredient/idMappings.csv', {})
mappingDict2 = mappingId('../ingredient/idMappings2.csv', {})
print(mappingDict1)
print(mappingDict2)
```

```python
def getRecipe(fileCnt):

    idArr = []
    recipeMat = []

    for i in range(fileCnt):

        print(i)

        with open(f'./recipeData/recipe_{i}.csv', encoding='utf-8') as recipe:

            reader = csv.reader(recipe)

            for line in reader:
                if (line == []) or (line[0] == 'id') or (line[1]=='') or (line[0] in idArr):
                    continue
                recipeMat.append(line[1].split('/'))
                idArr.append(line[0])
    
    print(idArr)
    print(recipeMat)

    return [idArr, recipeMat]


    
```

```python
ingIdArr = [] 

with open('./ingredientsData/ingredients_greater_than_5.csv', encoding='utf-8') as f:
    r = csv.reader(f)
    for ing in r:
        if ing[0] == 'id':
            continue
        ingIdArr.append(ing[0])

print(ingIdArr)
```

```python
recipeIdArr, recipeIngArr = getRecipe(3)
newRecipeIdArr = []
newRecipeIngArr = []

for i, recipeIng in enumerate(recipeIngArr):
    newRecipeIng = []
    for ing in recipeIng:
        if ing not in ingIdArr:
            continue
        else:
            try:
                newIng = mappingDict2[mappingDict1[ing]]
            except:
                continue
        if newIng not in newRecipeIng:
            newRecipeIng.append(newIng)
    
    if newRecipeIng == []:
        continue

    newRecipeIdArr.append(recipeIdArr[i])
    newRecipeIngArr.append(newRecipeIng)

print(newRecipeIdArr)
print(newRecipeIngArr)

    
```

```python
frequencyArr = list(map(lambda x: compress(oneHotEncoding(x)), newRecipeIngArr))
print(frequencyArr)
with open(f'./frequencyData/allFrequency_cleaned.csv', '+w', encoding='utf-8', newline='') as f:
    w = csv.writer(f)
    w.writerow(['id', 'frequency'])
    w.writerows(zip(newRecipeIdArr, frequencyArr))
```
