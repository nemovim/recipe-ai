```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
for i in range(1, 2):
    ingredientArr = []
    for seq in range(1+100*i, 101+100*i):
        url = f'https://10000recipe.com/bbs/ajax.html?q_mode=getMaterialContents&seq={seq}&recipe_seq=1'
        res = requests.get(url)
        name = res.json()['mname']
        print(seq, name)
        ingredientArr.append([seq, name])

    with open(f'ingredients_{i}.csv', 'w+', encoding='utf-8') as ing:
        writer = csv.writer(ing)
        writer.writerow(['id', 'name'])
        writer.writerows(ingredientArr)
```

```python
for j in range(147, 150):
    recipeArr = []

    for i in range(6900000+100*j, 6900101+100*j):
        print(i)

        url = f'https://10000recipe.com/recipe/{i}'
        html = requests.get(url).text

        if html.startswith('<script>'):
            continue

        anchorArr = BeautifulSoup(html, 'html.parser').select('#divConfirmedMaterialArea ul > a')

        ingredientArr = []

        for ing in anchorArr:
            ingredientArr.append(ing.get('href').split("'")[1])

        recipeArr.append([i, '/'.join(ingredientArr)])

    with open(f'recipe_{j}.csv', 'w+', encoding='utf-8', newline='') as recipe:
        writer = csv.writer(recipe)
        writer.writerow(['id', 'ingredients'])
        writer.writerows(recipeArr)
```
